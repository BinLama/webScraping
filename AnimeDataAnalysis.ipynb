```python
df[df["producers"].str.contains("Nonefound")] = np.nan
```

```python
unique_producers = []
for data in df["producers"]:
    if type(data) != float:
        new_data = data.split(";")
        print(new_data)
        for producer in new_data:
            if producer not in unique_producers:
                unique_producers.append(producer)
```

```python
unique_producers
```

```python
# create a dict of producers to aggregrate them
prod_dic = dict()
for i in unique_producers:
    prod_dic[i] = 0
    
for prod in unique_producers:
    for i in range(0,len(df["producers"])):
        if type(df["producers"][i]) != float and (df["producers"][i].__contains__(prod)) == True:
            prod_dic[prod] += 1
```

```python
prod_dic
```

```python
# https://stackoverflow.com/questions/22878743/how-to-split-dictionary-into-multiple-dictionaries-fast
# source for the help on using iterators and generators

def chunks(data, SIZE= 1000):
    it = iter(data)
    for i in range(0, len(data), SIZE):
        yield {k:data[k] for k in islice(it, SIZE)}
```

```python
for item in chunks(prod_dic, 10):
    print(item)
    data = item.items()
    x, y = zip(*data)
    print(x)
    break
```

```python
for item in chunks(prod_dic, 10):
    data = item.items()
    x, y = zip(*data)
    # plt.figure(figsize=(25.0,12.0), dpi=100)
    plt.bar(x,y, label="Company X Frequency")
    plt.xticks(rotation=90)
    plt.xlabel("Company")
    plt.ylabel("Frequency")
    plt.show()
```
